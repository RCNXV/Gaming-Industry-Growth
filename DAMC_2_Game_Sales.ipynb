#Data Cleaning


1. Change the datatype

  'Release' column wrong data type. it must in datetime64, not in object.

```python
gsc = game_sales.copy()
```

```python
gsc['Release'] = pd.to_datetime(gsc['Release'])
gsc.head()
```

```python
gsc.info()
```

2. Remove Duplicated Data

```python
gsc = gsc.drop_duplicates()
gsc.info()
```

3. Handling Missing Value

```python
gsc.isna().sum()
```

There is missing value in 'Series' column, we can see that because the game is in the first edition, so that they don't have series. We can change the missing value by its name of the game, instead of removing it.

```python
updated_df = gsc.copy()
updated_df['Series'].fillna(updated_df['Name'], inplace=True)

# Filter the rows where 'Series' has been replaced
replaced_rows = updated_df[updated_df['Series'] != gsc['Series']]

# Print the replaced rows
replaced_rows[['Name','Series']]
```

```python
gsc = updated_df
gsc.isna().sum()
```

```python
gsc
```

Change the data type in the 'Genre' column to a list, because there are several types of genres in certain games

```python
def transform_to_list(x):
    if ',' in x:
        return x.split(',')
    else:
        return [x]

gsc.loc[:,('Genre')] = gsc['Genre'].apply(lambda x: transform_to_list(x))
```
